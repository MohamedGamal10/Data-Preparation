```python
from bs4 import BeautifulSoup
with open('CMExport_LCAIE10015_Mohandseen_10.97.111.18_2020031405.xml', 'r') as f: 
    data = f.read()
Bs_data = BeautifulSoup(data, "lxml")
```

```python
from xml.dom import minidom
mydoc = minidom.parse('CMExport_LCAIE10015_Mohandseen_10.97.111.18_2020031405.xml')
MO = mydoc.getElementsByTagName('MO')
```

```python
import re
import pandas as pd
import time
r4=list()
data = {
        'Element':[],
        'Parameter':[],
        'Value':[]}

Main_DF=pd.DataFrame(data)

for Object in range(1,MO.length):
    Site_Element_Name = MO[Object].attributes['fdn'].value
    Site_Element_Data = Bs_data.find('mo', {'fdn':Site_Element_Name})
    r1=re.findall(r'<attr name="(.*?)</attr>',str(Site_Element_Data.find_all('attr')))
    fdn=r1[0].replace('fdn">','')
    del(r1[0])
    r2 = str(r1).split(',')
    
    for i in range (0,len(r2)):
        if '">' and not '=' in r2[i]:
            spliter=str(r2[i]).split('">')
            r4.append(spliter)
        elif '">' and '=' in r2[i]:
            spliter=str(r2[i]).split('=')
            r4.append(spliter)
        elif '=' and not '">' in r2[i]:
            spliter=str(r2[i]).split('=')
            r4.append(spliter)
    
    
    df=pd.DataFrame(r4,columns=['Para','Val'])
    df['Parameter']=df['Para'].str.replace("'", '').str.replace("[", '')
    df['Value']=df['Val'].str.replace("'", '').str.replace("]", '')
    df['Element']=fdn
    final_temp_DF=df.drop(['Para', 'Val'], axis=1)
    Main_DF=Main_DF.append(final_temp_DF,ignore_index=True)
    r4.clear()
```
